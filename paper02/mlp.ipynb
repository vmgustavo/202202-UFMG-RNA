```python
import torch
from torch import nn
from torch.utils.data import DataLoader

from skorch.toy import make_binary_classifier
from skorch import NeuralNetBinaryClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

import numpy as np

from src.datasets import get_blobs
```

```python
data, target = get_blobs(n_obs=128)
target = target > 0
X_train, X_test, y_train, y_test = train_test_split(data.values, target.values)
```

```python
classifier = make_binary_classifier(
    input_units=2, output_units=1,
    hidden_units=256, num_hidden=1,
    nonlin=nn.Tanh(), output_nonlin=nn.Tanh(),
)
```

```python
net = NeuralNetBinaryClassifier(
    classifier,
    max_epochs=256,
    lr=0.1,
    device="cpu",
)
```

```python
net.fit(X_train.astype(np.float32), y_train.astype(np.float32));
```

```python
y_pred = net.predict(X_test.astype(np.float32)) > 0 
```

```python
print(classification_report(y_pred=y_pred, y_true=y_test))
```

```python
class Perceptron(torch.nn.Module):
    def __init__(self):
        super(Perceptron, self).__init__()
        self.fc = nn.Linear(1,1)
        self.relu = torch.nn.ReLU() # instead of Heaviside step fn

    def forward(self, x):
        output = self.fc(x)
        output = self.relu(x) # instead of Heaviside step fn
        return output
```
